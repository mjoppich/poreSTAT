```python
def getDirTypes(graph, mirna):

    nodeData = graph.nodes[mirna]["attr_dict"]

    nodeDir = nodeData.get("node_expr_direction", "N/A")

    if nodeDir == "N/A":
        return {}, {}

    nodeNN = [x for x in graph.neighbors(mirna)]

    dirTypes = Counter()
    nnDir = Counter()

    for nn in nodeNN:
        tgt = nn
        edgeData = graph.edges[(mirna, nn)]
                
        dtype = edgeData.get("dir_type", "").split("_")[0]
        
        nnDirType = graph.nodes[nn]["attr_dict"].get("node_expr_direction", "N/A")
        nnDir[ nnDirType ] += 1
        
        dirTypes[dtype] += 1
        

    return dirTypes, nnDir

getDirTypes(mgGraph, "miR-27")
```
